```python
import pandas as pd

data_bersih = []
log_perubahan_simbol = []

# 1. BACA DATA DAN CLEANING SIMBOL
with open('Data tiktok.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

for line in lines[1:]:
    line = line.strip()
    if not line: continue

    parts = line.split(',', 1)

    if len(parts) == 2:
        user = parts[0].strip()
        komen_asli = parts[1].strip()

        # --- PROSES CLEANING (Preprocessing Tahap 1) ---
        komen_baru = komen_asli.rstrip(';')

        if user.startswith('"') and user.endswith('"'): user = user[1:-1]
        if komen_baru.startswith('"') and komen_baru.endswith('"'):
            komen_baru = komen_baru[1:-1]

        komen_baru = komen_baru.replace('""', '"')


        if komen_asli != komen_baru:
            log_perubahan_simbol.append({
                'Username': user,
                'Sebelum (Kotor)': komen_asli,
                'Sesudah (Bersih)': komen_baru
            })

        data_bersih.append({'Username': user, 'Komentar': komen_baru})

# Buat DataFrame Utama
df = pd.DataFrame(data_bersih)

# 2. DETEKSI DUPLIKAT
data_duplikat = df[df.duplicated(subset=['Komentar'], keep='first')]

# 3. HAPUS DUPLIKAT (Finalisasi)
df_final = df.drop_duplicates(subset=['Komentar'], keep='first')

# --- MENAMPILKAN HASIL ---
print("LAPORAN DATA CLEANING (PREPROCESSING TAHAP 1)")
print(f"Total Data Awal Terbaca  : {len(df)}")
print(f"Data Mengalami Perbaikan : {len(log_perubahan_simbol)} baris (Simbol ;; atau kutip dibuang)")
print(f"Data Duplikat Dihapus    : {len(data_duplikat)} baris")
print(f"Total Data Akhir (Bersih): {len(df_final)} baris")

print("\nDATA YANG DIBERSIHKAN SIMBOLNYA (Before vs After):")
df_log = pd.DataFrame(log_perubahan_simbol)
if not df_log.empty:
    print(df_log[['Sebelum (Kotor)', 'Sesudah (Bersih)']].head(10).to_string(index=False))
else:
    print("Tidak ada simbol yang perlu dibersihkan.")

print("\n DATA DUPLIKAT YANG DIBUANG:")
if not data_duplikat.empty:
    print(data_duplikat[['Username', 'Komentar']].head(5).to_string(index=False))
else:
    print("Tidak ada duplikat ditemukan.")

# Simpan File Akhir
df_final.to_excel('Scrap Cleaning.xlsx', index=False)
print("\n\n[INFO] File bersih tersimpan sebagai 'Scrap Cleaning.xlsx'")
```
