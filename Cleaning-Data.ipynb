```python
import pandas as pd

# 1. LOAD DATA
nama_file = 'Data Scrap.csv'
df = pd.read_csv(nama_file, encoding='utf-8-sig')

print(f"Total Data: {len(df)} baris")
```

```python
#NOISE REMOVAL

# 2. HAPUS DUPLIKAT
df_clean = df.drop_duplicates(subset=['Komentar'], keep='first')

# 3. HAPUS DATA KOSONG (NaN)
df_clean = df_clean.dropna(subset=['Komentar'])

# 4. HAPUS KOMENTAR TIDAK BERMAKNA
df_clean = df_clean[df_clean['Komentar'].str.len() > 3]

# CEK HASIL
jumlah_awal = len(df)
jumlah_akhir = len(df_clean)
jumlah_dibuang = jumlah_awal - jumlah_akhir

print(f"Data Awal      : {jumlah_awal}")
print(f"Data Dibuang   : {jumlah_dibuang} (Duplikat/Spam)")
print(f"Data Tersisa   : {jumlah_akhir} (Siap Labeling)")

# 5. SIMPAN KE EXCEL
nama_file_siap_label = 'Scrap to Label.xlsx'
df_clean.to_excel(nama_file_siap_label, index=False)
```

```python
#BUAT FILE SIAP LABELING
# 1. Acak urutan baris (Shuffle)
df_shuffled = df_clean.sample(frac=1, random_state=42).reset_index(drop=True)

# 2. Ambil 1.500 data pertama hasil acakan
df_sample = df_shuffled.head(1500)

# 3. Simpan ini untuk dilabeli
df_sample.to_excel("Bahan-Labeling.xlsx", index=False)

print(f"{len(df_clean)} data 1.500 sampel acak untuk dilabeli.")
```
