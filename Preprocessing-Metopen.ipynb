```python
#Load Dataset
import pandas as pd
import io
import os

filename = 'Scraping.csv'
print(f"Berhasil load dataset '{filename}'")

cleaned_rows = []
try:
    with open(filename, 'r', encoding='utf-8', errors='replace') as f:
        lines = f.readlines()

    for line in lines:
        line = line.strip()
        # Hapus titik koma berlebih di akhir
        while line.endswith(';'):
            line = line[:-1]
        # Perbaiki format tanda kutip
        if len(line) > 1 and line.startswith('"') and line.endswith('"'):
            content = line[1:-1].replace('""', '"')
            cleaned_rows.append(content)
        else:
            cleaned_rows.append(line)

    # Buat DataFrame
    cleaned_csv_str = "\n".join(cleaned_rows)
    df = pd.read_csv(io.StringIO(cleaned_csv_str), sep=',')

    print(f" Jumlah Data Awal: {len(df)} baris")
    print("Contoh Data:")
    display(df.head(10))

except Exception as e:
    print(f"‚ùå Error saat membaca file: {e}")
```

```python
#Hapus Data Duplicate
print("üîç Memeriksa duplikasi tweet...")
jumlah_awal = len(df)

df = df.drop_duplicates(subset=['full_text'], keep='first').reset_index(drop=True)

jumlah_akhir = len(df)
jumlah_dibuang = jumlah_awal - jumlah_akhir

if jumlah_dibuang > 0:
    print(f"‚ö†Ô∏è Ditemukan dan dihapus {jumlah_dibuang} tweet duplikat.")
else:
    print("‚úÖ Tidak ada duplikat ditemukan.")

print(f"üìä Total Data Bersih: {jumlah_akhir} baris")
```

```python
#Case Folding (Agar teks kecil semua)
print("üî† Melakukan Case Folding...")

# Simpan teks asli untuk perbandingan (opsional)
df['original_text'] = df['full_text']

# Ubah ke lowercase
df['processed_text'] = df['full_text'].str.lower()

print("‚úÖ Case Folding Selesai.")
print(df[['original_text', 'processed_text']].head(10))
```

```python
#Menghapus noise
import re

def clean_noise(text):
    text = re.sub(r'http\S+', '', text)       # Hapus URL
    text = re.sub(r'@\w+', '', text)          # Hapus Mention
    text = re.sub(r'#\w+', '', text)          # Hapus Hashtag
    text = re.sub(r'\d+', '', text)           # Hapus Angka
    text = re.sub(r'[^\w\s]', ' ', text)      # Hapus Simbol/Tanda Baca
    text = re.sub(r'\s+', ' ', text).strip()  # Hapus Spasi Berlebih
    return text

print("üßπ Membersihkan noise (URL, @, #, simbol)")
df['processed_text'] = df['processed_text'].apply(clean_noise)

print("‚úÖ Noise Removal Selesai.")
print(df[['processed_text']].head(10))
```

```python
#Tokenizing
import nltk
from nltk.tokenize import word_tokenize

# Download resource NLTK
nltk.download('punkt')
nltk.download('punkt_tab')

print("‚úÇÔ∏è Melakukan Tokenizing...")
df['tokens'] = df['processed_text'].apply(word_tokenize)

print("‚úÖ Tokenizing Selesai.")
print(df[['tokens']].head(3))
```

```python
#Mengubah Kata Tidak Baku
kamus_slang = {
    'yg': 'yang', 'gak': 'tidak', 'bgt': 'banget', 'tdk': 'tidak',
    'kalo': 'kalau', 'krn': 'karena', 'jgn': 'jangan', 'sdh': 'sudah',
    'aja': 'saja', 'dgn': 'dengan', 'utk': 'untuk', 'mbg': 'makan bergizi gratis',
    'gw': 'saya', 'aku': 'saya', 'tak': 'tidak'
}

def normalisasi(token_list):
    return [kamus_slang.get(word, word) for word in token_list]

print("üìñ Menormalkan kata slang...")
df['tokens'] = df['tokens'].apply(normalisasi)

print("‚úÖ Normalisasi Selesai.")
# Cek sampel
print(df['tokens'].head(10))
```

```python
#Menghapus kata yang tak penting
!pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

print("üóëÔ∏è Menghapus Stopwords...")

factory = StopWordRemoverFactory()
stopwords_indo = set(factory.get_stop_words())
# Tambahan stopword manual
stopwords_indo.update(['di', 'ke', 'dari', 'yang', 'ini', 'itu', 'dan'])

def remove_stopwords(token_list):
    return [word for word in token_list if word not in stopwords_indo]

df['tokens'] = df['tokens'].apply(remove_stopwords)

print("‚úÖ Stopword Removal Selesai.")
print(df['tokens'].head(10))
```

```python
#Menyimpan Hasil Bersih
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tqdm import tqdm # Untuk progress bar

# Aktifkan progress bar pandas
tqdm.pandas()

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming_process(token_list):
    kalimat = ' '.join(token_list)
    stemmed = stemmer.stem(kalimat)
    return stemmed.split() # Kembalikan sebagai list, atau hapus .split() jika ingin kalimat utuh

print("‚è≥ Sedang melakukan Stemming (Mohon tunggu)...")
df['final_tokens'] = df['tokens'].progress_apply(stemming_process)

# Gabungkan tokens kembali jadi kalimat utuh untuk disimpan
df['clean_text'] = df['final_tokens'].apply(lambda x: ' '.join(x))

# Simpan ke CSV baru
output_file = 'Data_Bersih_MBG.csv'
df.to_csv(output_file, index=False)

print("\n‚úÖ SEMUA PROSES SELESAI!")
print(f"üìÅ File bersih tersimpan sebagai: {output_file}")
display(df[['full_text', 'clean_text']].head(10))
```
